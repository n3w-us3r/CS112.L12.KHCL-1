# TẢO BIỂN

Tảo biển sinh sản rất nhanh khi có môi trường thuận lợi với chúng và có những loài còn tiết ra môi trường những chất độc hại.
Một loại tảo nâu trong môi trường nước bị ô nhiễm nặng sinh sản theo quy luật sau:

Ngày đầu tiên (ngày 0) có n cá thể ở mức 1,
Ở mỗi ngày tiếp theo, mỗi cá thể mức i sinh ra i cá thể mức 1, các cá thể mới sinh sẽ sinh sôi, phát triển từ ngày hôm sau.
Bản thân các cá thể mức i phát triển thành mức i+1 và chu kỳ phát triển trong ngày chấm dứt.
Hãy xác định sau k ngày trong nước biển có bao nhiêu cá thể.

Dữ liệu: Vào từ thiết bị nhập chuẩn gồm một dòng chứa 2 số nguyên n và k (1 ≤ n ≤ 1000, 1 ≤ k ≤ 1018).

(test hiện tại chỉ có k <= 105,  thầy sẽ chấm lại sau khi nâng cấp test)

Kết quả: Đưa ra thiết bị xuất chuẩn một số nguyên – số lượng cá thể  tảo theo mô đun 109+7.

Ví dụ:

INPUT: 3 2

OUTPUT: 15


#1.Abstraction 

Find **n'th term** of a sequence  $u_1$ , $u_2$  , $u_3$  ,… $u_{n-2}$  , $u_{n-1}$,$u_n$ defined below : 

$u_1$=k 

$u_2$=2k    

$u_n$=3.$u_{n-1}$ - $u_{n-2}$

  n , k $\epsilon$ Z

###Constraints

n is an integer greater than or equal to 1 and less or equal to 1000 ( 1 ≤ n ≤ 1000 , n ∈ Z )

k is an integer greater than or equal to 1 and less or equal to  $10^{18}$ (1 ≤ k ≤ $10^{18}$ , k ∈ Z)

#2. Pattern Recognition: 
Divide and conquer 

#3. Algorithm Design 

 **Input Format**

The only line contains space-separated integers, n and k, respectively ( 1 ≤ n ≤ 1000 , 1 ≤ k ≤ $10^{18}$)

 **Output Format**

The only line contains a single integer as the result of problem.

**Solution**

Let's say n is the initial seaweed number, k is the day we would like to know the number of seaweed available at that day and x is the number of seaweed at k-th day

For example, with  n = 1, we will have an integer sequence:

```
  |  n  |  k  |  x  | 
  ------------------
  |  1  | 0   | 1   | 
  |  1  | 1   | 2   | 
  |  1  | 2   | 5   |  
  |  1  | 3   | 13  | 
  |  1  | 4   | 34  | 
  |  1  | 5   | 89  | 
  |  1  | 6   | 233 | 
  |  1  | 7   | 610 | 
  ...
```
Take a look at the above table, we can come up with a formula by looking for the sequence 1, 2, 5, 13, 34, 89, 233, 610 on google 

```
u(k) = 3 * u(k - 1) - u(k - 2) with u(0) = 1, u(1)=2, k >= 2
```

With the above formula, we can find that it's very familiar because it remind us of k-th fibonacci formula. We can come up with a matrix that will help us caculate the n-th number in the sequence.

```
[ [3 -1]
  [1  0] ]
```

Here's how we get the above matrix:

```
Fisrtly, we are assuming a linear recurrence equation like below:

  u(k) = a * u(k - 1) + b * u(k - 2) (1) 
  
  where a,b are constants

Now we try to represent the equation (1) in terms of matrix:

  | u(k)   | = Matrix 'X' * | u(k-1) |
  | u(k-1) |                | u(k-2) |

The dimension of the first matrix is 2 x 1 and the last matrix is also 2 x 1, so the dimension of 'X' must be 2 x 2

So according to our equation:
u(k) = a * u(k-1) + b * u(k-2)
u(k-1) = u(k-1)

'X' would be: | a b |
              | 1 0 |

Now we have:
  | u(k)   | = | a b |  *  | u(k-1) |
  | u(k-1) |   | 1 0 |     | u(k-2) |

As we found the equation with correct a and b to be:
  u(k) = 3 * u(k-1) - u(k-2)

So, a will be 3 and b will be -1
We can come to the solution:
  Matrix 'X' = | 3 -1|
               | 1 0 |

```

So, for this problem we can use our knowledge about matrix and divide and conquer to access this problem, similar to how we solve the n-th fibonacci problem. Why don't we just do a simple O($2^k$) recursive solution or O(k) dynamic programming solution instead ? As the statement mentioned, k can be up to 1e18 so O($2^k$) approach and O(k) approach won't help us get there. 

So we need to calculate the matrix power k using divide and conquer strategy, our final answer will be the sum of two number on the first row of the result matrix. <br>

```
Base cases:
0 < k < 2 => F(k) = 1 
```

Time complexity: O(log(k)) <br>
Space complexity: O(1) or O(log(n)) if we consider the stack size of the recursive function call



# Pseudocode:

```
# initialize

matrix = [ [3, -1], [1, 0] ]
CONST = int(1e9 + 7)

# calculate the matrix power k using divide and conquer strategy 
# our final answer will be the sum of two number 
# on the first row of the result matrix 

func pow(base_matrix, exponent):
  if exponent == 1:
    return (base_matrix[0][0] + base_matrix[0][1]) % CONST

  pow(base_matrix, int(exponent/2))
  multiply(base_matrix, base_matrix)

  if exponent % 2 != 0:
    multiply(base_matrix, matrix)

  return (base_matrix[0][0] + base_matrix[0][1]) % CONST

func solve(n, k):
  if k < 2:
    print(n * 2)
    return

  print((pow(matrix, k) * n) % CONST)


```

````

matrix = [ [3, -1], [1, 0] ]
CONST = int(1e9 + 7)

def cloning(old_list):
  new_list = [ x[:] for x in old_list ]
  return new_list

def multiply(_matrix, temp):
  global CONST

  i = 0

  for row in _matrix:
    a = row[0]
    b = row[1]
    _matrix[i][0] = ((a * temp[0][0]) + (b * temp[1][0])) % CONST 
    _matrix[i][1] = ((a * temp[0][1]) + (b * temp[1][1])) % CONST
    i += 1

def pow(base_matrix, exponent):
  global matrix
  global base_case
  global CONST

  if exponent == 1:
    return (base_matrix[0][0] + base_matrix[0][1]) % CONST

  pow(base_matrix, int(exponent/2))
  multiply(base_matrix, cloning(base_matrix))

  if exponent % 2 != 0:
    multiply(base_matrix, matrix)

  return (base_matrix[0][0] + base_matrix[0][1]) % CONST 

def solve(n, k):
  global matrix

  if k < 2:
    print(n * 2)
    return

  temp = [ x[:] for x in matrix]

  print((pow(temp, k) * n) % CONST)

n, k = [ int(x) for x in input().split() ]

solve(n, k)
````
