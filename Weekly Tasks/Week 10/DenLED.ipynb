### ĐÈN LED

Alice vừa hoàn thành một hợp đồng lắp bảng hiển thị số bằng đèn LED. Mỗi chữ số được hiển thị trong một khung chữ nhật với 7 ống đèn LED. Bằng cách bật các ống đèn LED thích hợp ta có thể hiển thị chữ số bất kỳ. Số đèn sáng càng nhiều thì việc hiển thị chữ số đó càng tốn năng lượng. Ví dụ, hiển thị chữ số 9 sẽ tốn năng lượng hơn hiển thị chữ số 7.

image.png

Sau khi bàn giao sản phẩm trong tay Alice còn thừa lại một số khá nhiều các khung hiển thị số và một cục pin nguồn. Dung lượng pin cho phép bật sáng n ống đèn LED. Alice muốn dùng pin bật đúng n ống đèn để hiển thị một số và số hiển thị được phải có tổng chữ số là lớn nhất.

Hãy xác định tổng lớn nhất của các chữ số của số có thể bật sáng.

Dữ liệu: Vào từ thiết bị nhập chuẩn gồm một dòng chứa số nguyên n (2 ≤ n ≤ $10^6$).

Kết quả: Đưa ra thiết bị xuất chuẩn một số nguyên – tổng lớn nhất đạt được.

INPUT | OUTPUT
--|--
7 | 11


## 1. Abstraction

Có một số  nguyên là mức năng lượng, phải tìm cách tận dụng số năng lượng đó để thắp sáng các bóng led (mỗi bóng led là một nét biểu diễn chữ số) tạo thành 1 chuỗi các chữ số  sao cho tổng các chữ số là lớn nhất.

## 3. Algorithm Design

Ta có 1 bảng các chữ số được biểu diễn sử dụng các bóng led, mỗi bóng led là 1 nét trong số các nét cấu thành 1 chữ số

image.png

Qua nhìn nhận, có thể thấy ta có trường hợp với số nét là ```n``` (n<=7) ta có thể biểu diễn được nhiều chữ số khác nhau, ví dụ với số nét là ```6``` thì có thể biểu diễn chữ số  ```6``` hoặc ```9```, số nét là ```5``` có thể biễu diễn chữ số  ```2``` hoặc ```3``` hoặc ```5```, ...
Nhưng theo yêu cầu bài toán có thể thấy, ta cần biểu diễn được một chuỗi các chữ số từ ```n``` nét sao cho tổng các chữ số phải là lớn nhất, vậy các với ```n``` (n<=7) ta chỉ chọn chữ số lớn nhất có thể được biểu diễn bởi ```n``` nét. Ta có thể lập được bảng ánh xạ sau:

số nét | chữ số
--|--
2 | 1
4 | 4
5 | 5
3 | 7
7 | 8
6 | 9

Tuy nhiên, chưa dừng lại ở đó. Qua nhìn nhận ở bảng ánh xạ vừa rút được, có thể thấy với số nét là ```6```, thay vì biểu diễn 1 chữ số  ```9``` thì ta có 1 cách biểu diễn khác là biễu diễn 2 chữ số  ```7``` (1 chữ số ```7``` cần 3 nét để biểu diễn, và ta có 6 nét). Qua ví dụ trên có thể thấy cách thứ 2 mới là cách tận dụng ```6``` nét đúng đắn. Vì vậy, với những số nét như ```6``` thì chưa đủ nhỏ để  làm số nét cơ sở, cần phải được lược đi. Vậy ta có bảng ánh xạ mới như sau:

số nét | chữ số
--|--
2 | 1
4 | 4
3 | 7

Tới đây, bài toán đã đơn giản hơn rất nhiều. Có thể thấy chữ số lớn nhất là chữ số ```7``` với ```3``` nét biểu diễn, vậy ta sẽ ưu tiên biểu diễn nhiều chữ số ```7``` nhất có thể, kế đến tùy thuộc vào số nét còn sót lại sau khi đã biểu diễn nhiều chữ số ```7``` nhất có thể, ta quyết định xem sẽ biểu diễn chữ số ```4``` hay ```1```. Nếu còn sót 1 nét thì sẽ là chữ số ```4```, nếu sót 2 nét thì sẽ là chữ số ```1```

```
n = int(input())

digit_7_num = n div 3

remainder = n mod 3

if(remainder == 0) then print(digit_7_num * 7)
else if(remainder == 1) then print((digit_7_num - 1) * 7 + 4)
else if(remainder == 2) then print(digit_7_num * 7 + 1)

```

## 4. Implementation

````
n = int(input())

digit_7_num = n // 3

remainder = n % 3

if(remainder == 0):
    print(digit_7_num * 7)
elif(remainder == 1):
    print((digit_7_num - 1) * 7 + 4)
elif(remainder == 2):
    print(digit_7_num * 7 + 1)
````
