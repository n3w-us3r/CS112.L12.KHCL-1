# Xâu Con Tương Đương

Cho xâu s chỉ chứa các ký tự la tinh thường độ dài n không quá 105. Ký hiệu s[i..j] là xâu con các ký tự liên tiếp nhau của s từ vị trí i đến vị trí j. Hai xâu con x = s[a..b] và y = s[c..d] được gọi là cùng một lớp tương đương nếu có cách đổi chỗ các ký tự trong một xâu để nhận được xâu kia. Các vị trí được đánh số từ 1 trở đi.

Ví dụ với s = “abcbacaac”, s[1..3] tương đương với s[4..6], còn s[1..3] không tương đương với s[5..7].

Cho q truy vấn, mỗi truy vấn là 4 số nguyên a, b, c, d xác định các xâu con x và y. Với mỗi truy vấn hãy xác định hai xâu con này có cùng một lớp tương đương hay không và đưa ra câu trả lời tương ứng “YES” hoặc “NO”.

Dữ liệu: Vào từ thiết bị nhập chuẩn:

Dòng đầu tiên chứa xâu s,

Dòng thứ 2 chứa số nguyên q (1 ≤  q  ≤ $10^5$),

Mỗi dòng trong q dòng tiếp theo chứa 4 số nguyên a, b, c, d (1 ≤ a ≤ b ≤ n, 1 ≤ c ≤ d ≤ n).

Kết quả: Đưa ra thiết bị xuất chuẩn các câu trả lời xác định được cho mỗi truy vấn, mỗi câu trả lời trên một dòng.

Dữ liệu mẫu:

INPUT | OUTPUT
    --|--
2      | YES
1 3 4 6| NO
1 3 5 7|

## 1. Abstraction
Với cặp chuỗi con a, b thuộc chuỗi mẹ s. Xét xem 2 chuỗi có cùng số tần suất xuất hiện của mỗi kí tự trong 2 chuỗi hay không.

Input Format <br>

>Dòng đầu tiên chứa xâu s <br>
Dòng thứ 2 chứa số nguyên q (1 ≤  q  ≤ $10^5$) <br>
Mỗi dòng trong q dòng tiếp theo chứa 4 số nguyên a, b, c, d (1 ≤ a ≤ b ≤ n, 1 ≤ c ≤ d ≤ n). <br>


Output Format <br>
>Một số nguyên n là kết quả của bài toán

## 2. Decomposition
Dựa vào đề bài, 2 chuỗi con tương đương nhau là khi nếu có cách đổi chỗ các ký tự trong một xâu để nhận được xâu kia. Vậy để xác định liệu có cách đổi chỗ nào để biến 1 chuỗi thành chuỗi còn lại hay không, ta nhận thấy rằng chỉ cần 2 chuỗi có cùng độ dài, cùng số loại kí tự và tần suất của mỗi loại kí tự là như nhau thì chắc chắc luôn có cách hoán đổi để biến chuỗi này thành chuỗi kia. Vậy ta chỉ cần kiểm tra độ dài 2 chuỗi và kiểm tra tần xuất của mỗi loại kí tự trong 2 chuỗi rồi so sánh với nhau là có thể kết luận 2 chuỗi có tương đương hay không. Vậy bài toán lúc này là xác định xem liệu 2 chuỗi có được biểu diễn bởi những loại kí tự giống nhau với số tần xuất mỗi loại kí tự tương đương nhau hay không ?

## 3. Pattern Recognition

prefix sum, hash, counting sort

## 4. Algorithm Design

Để có thể đếm được tần xuất của mỗi loại kí tự trong chuỗi, ta có thể sử dụng phương pháp đếm phân phối với 26 loại kí tự. Tuy nhiên, ta cần xác định tần xuất của mỗi loại chữ cái cho các chuỗi con được yêu cầu. 
nếu mỗi lần xuất hiện yêu cầu lại thực hiện việc duyệt tuyến tính để đếm thì sẽ làm độ phức tạp đạt tới mức O($n^2$), như vậy sẽ không đạt được yêu cầu về giới hạn thời gian. Vì vậy ta có thể áp dụng phương pháp tổng tiền tố để xác định được tần xuất mỗi loại kí tự trong các chuỗi con 1 cách hiệu quả. Bằng cách tại mỗi vị trí i (0 <= i < len(s)), ta lưu trữ tần xuất của mỗi loại kí tự từ trong chuỗi con từ vị trí 0 đến i. Vậy để xác định tần xuất mỗi loại kí tự của chuỗi con [a, b] bất kì, ta lấy tần xuất mỗi loại chữ cái tính từ 0 đến b trừ đi tần xuất mỗi loại chữ cái tính từ 0 đến a-1 (a >= 1). 

>Time complexity: O(26 * n)

>Space complexity: O(26 * n)

```
# Hình thành mảng tổng tiền tố 
def cnt(s: string) -> array:
  n = length of s
  prefix_sum_arr = an array of n sub-array with initial length of 26

  prefix_sum_arr[0][ascii(s[0])-97) = 1

  for i from 0 to n-1:
    for j from 0 to 25:
      prefix_sum_arr[i][j] = prefix_sum_arr[i-1][j] if j != ascii(s[i])-97 else prefix_sum_arr[i-1][j] + 1

  return prefix_sum_arr

def substr_char_anagram(s: string, a: int, b: int, prefix_sum: arr) -> array:
    ans = an array of 26 elements with value of 0

    for i from 0 to 26:
        ans[i] = prefix_sum[b-1][i] - prefix_sum[a-1][i]
    ans[ascii(s[a-1])-97] += 1

    return ans


def solve() -> void:

    q = input(int)

    prefix_sum = cnt(s)

        a, b, c, d = input(an array of four integers)

        if a == c and b == d:
            write('YES\n')
            continue
        elif b - a != d - c:
            write('NO\n')
            continue

```

Tuy nhiên, vẫn còn 1 phương pháp tối ưu hơn. Ta có thể tìm cách hash 2 chuỗi con a và b thành key1 và key2 tương ứng, sau đó tiến hành so sánh key1 và key2 nếu trùng nhau thì có thể kết luận 2 chuỗi tương đương nhau. Để hash, ta vẫn dựa trên yếu tố các chữ cái cấu thành chuỗi 1 có giống với các chữ cái cấu thành chuỗi 2 hay không ? (không quan trọng thứ tự các chữ cái). Ta gọi tổng mã ascii của các kí tự trong 2 chuỗi lần lượt là key1, key2.
>Tuy nhiên với trường hợp "ac" và "bb": <br>
key1 = 97 + 99 = 196 <br>
key2 = 98 + 98 = 196

Vậy có thể thấy key1 và key2 cùng bằng 196 nhưng "ac" và "bb" không tương đương. Vậy công thức hash không thể chỉ đơn giản là tính tổng mã ascii của tất cả kí tự trong chuỗi. Vậy nếu ta lũy thừa mỗi mã ascii cho 1 số n rồi tính tổng tất cả thì có thể sẽ hash được key đạt yêu cầu về tính duy nhất. Qua thực nghiệm với số mũ tăng dần từ 2, thuật toán pass hết bộ test với số mũ bằng 3. Để xác định được tổng số ascii của tất cả kí tự trong chuỗi con [a, b], ta vẫn sẽ dùng phương pháp tổng tiền tố.
Tại mỗi vị trí i (0 <= i < len(s)), ta sẽ lưu trữ tổng của tất cả các kí tự quy đổi sang ascii và lũy thừa cho 3 từ vị trí 0 đến i. 
Vậy ta có công thức tính key của chuỗi con [a, b]:
> key = key[b] - key[a-1] (a > 0)

>Time complexity: O(n) <br>
>Space complexity: O(n)

```
# hình thành mảng tổng tiền tố
def cnt(s: string, prefix_sum: array):
    prefix_sum[0] = pow(ascii(s[0]), 3) 

    for i from 1 to length(s)-1:
        prefix_sum[i] = prefix_sum[i-1] + pow(ascii(s[i]), 3)


def solve():
    s = input(string)

    q = input(int)

    prefix_sum = an array of length(s) elements with value of 0
    cnt(s, prefix_sum)

    for i from 0 to q-1:
        a, b, c, d = input(an array of four integers)

        if a == c and b == d:
            write('YES\n')
            continue
        elif b - a != d - c:
            write('NO\n')
            continue

        if prefix_sum[b-1] - prefix_sum[a-1] + pow(ascii(s[a-1]), 3) == prefix_sum[d-1] - prefix_sum[c-1] + pow(ascii(s[c-1]), 3):
            write('YES\n')
        else:
            write('NO\n')


solve()
```


## 5. Implentation

Cách 1: prefix sum + counting sort

````
from sys import stdin, stdout


def cnt(s, prefix_sum):
    prefix_sum[0][ord(s[0])-97] = 1

    for i in range(1, len(s)):
        for j in range(26):
            prefix_sum[i][j] = prefix_sum[i-1][j] if j != ord(
                s[i])-97 else prefix_sum[i-1][j] + 1


def substr_char_anagram(s, a, b, prefix_sum, ans):
    for i in range(26):
        ans[i] = prefix_sum[b-1][i] - prefix_sum[a-1][i]

    ans[ord(s[a-1])-97] += 1


def solve():
    s = stdin.readline().strip()

    q = int(stdin.readline().strip())

    prefix_sum = [[0] * 26 for i in range(len(s)+1)]
    cnt(s, prefix_sum)

    a_b = [0] * 26
    c_d = [0] * 26
    for i in range(q):
        a, b, c, d = [int(n) for n in stdin.readline().split()]

        if a == c and b == d:
            stdout.write('YES\n')
            continue
        elif c - a != d - b:
            stdout.write('NO\n')
            continue

        substr_char_anagram(s, a, b, prefix_sum, a_b)
        substr_char_anagram(s, c, d, prefix_sum, c_d)

        if a_b == c_d:
            stdout.write('YES\n')
        else:
            stdout.write('NO\n')


solve()
````
